```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df= pd.read_csv('fordgobike2018.csv')
df.shape
```

```python
#Assessing data 
df
```

##  User Types 

> According to the website, membership prices is cheaper than single rides. https://www.lyft.com/bikes/bay-wheels/pricing

```python
user_types = df['user_type'].value_counts()
plt.pie(user_types, labels = user_types.index, explode=(0, 0.2), autopct='%1.1f%%', startangle = 90,
        counterclock = False, colors= ['y', 'g']);
plt.title('customer Vs Subscriber');
```

## gender of members

> Males 69%
> Females 23.5% 

```python
base_color = sb.color_palette()[0]
sb.countplot(data = df, y = 'member_gender', color = base_color);
plt.xticks(rotation = 90);
plt.ylabel('Gender', fontsize= 14);
plt.xlabel('Number of Riders', fontsize= 14);
plt.title('gender of members', fontsize= 22);
```

## Age of members
> average of age 34 years

```python
df['age'] = 2019 - df.member_birth_year 
plt.figure(figsize = [14, 4]) # larger figure size for subplots
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bin_edges = np.arange(0, df['age'].max()+4, 4)
plt.hist(data = df, x = 'age', bins = bin_edges, rwidth = 0.7);
plt.title(' Age of Member', fontsize = 22);
plt.xlabel('Age');
plt.ylabel('Number of Members');
plt.xlim(15, 80)
```

## Rides Per Months
 > weather has influence on trips 



```python
## https://stackoverflow.com/questions/25146121/extracting-just-month-and-year-from-pandas-datetime-column
df['month'] = pd.DatetimeIndex(df['start_time']).month
plt.figure(figsize=[8 ,4]);
sb.countplot(data = df, x = 'month', color = base_color);
plt.title(' rides per month', fontsize = 22);
plt.xlabel('months');
plt.ylabel('Rides');
```

## Duration in Minutes
> Average of Duration = 14 minutes

```python
df['duration_min'] = df['duration_sec'] / 60  
bin_edges = np.arange(0, df['duration_min'].max()+2, 2)
plt.hist(data = df, x = 'duration_min', bins = bin_edges);
plt.xlim(0, 60);
plt.title('Duration in Minutes', fontsize = 22);
plt.xlabel('Duration in Minutes');
```

## Member Age and Gender
 - Average age of Females = 35
 - Average age of Males = 37

```python
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]

sb.violinplot(data = df, x = 'age', y = 'member_gender', color = base_color,
              inner = None)
plt.xlim(18, 70)
plt.title('Member Age and Gender', fontsize = 22);
```

## User Type and Gender

```python
sb.countplot(data = df, x = 'user_type', hue = 'member_gender')
plt.title('User Type and Gender', fontsize = 22);
```

## Duration in Minutes and Gender

***females ride for long duration***

```python
plt.figure(figsize = [10, 5])
plt.subplot(2, 1, 2)
sb.boxplot(data = df, x = 'duration_min', y = 'member_gender', color = base_color)
plt.xlim(0, 30); # set y-axis limits to be same as left plot
plt.title('Duration in Minutes and Gender', fontsize = 22);
```

## Duration in Minutes per month
> Members got long trip in Summer

```python
plt.figure(figsize = [10, 10])
sb.regplot(data = df, x = 'month', y = 'duration_min', fit_reg = False,
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/10})
plt.ylim(0, 1500)
plt.title('Duration in Minutes per month', fontsize = 22);
```

## Who is ridind for longer trip
**Females got the longest trip**

```python
cat_means = df.groupby(['member_gender', 'user_type']).mean()['duration_min']
cat_means = cat_means.reset_index(name = 'duration_min_avg')
cat_means = cat_means.pivot(index = 'user_type', columns = 'member_gender',
                            values = 'duration_min_avg')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'mean(duration in minutes)'})
```

```python
def mean_poly(x, y, bins = 5, **kwargs):
    """ Custom adapted line plot code. """
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    means = y.groupby(data_bins).mean()

    # create plot
    plt.errorbar(x = bin_centers, y = means, **kwargs)

bin_edges = np.arange(0.25, df['month'].max()+1,1)
g = sb.FacetGrid(data = df, hue = 'member_gender', size = 5)
g.map(mean_poly, "month", "duration_min", bins = bin_edges)
g.set_ylabels('mean(duration in minutes)')
g.add_legend()
```
